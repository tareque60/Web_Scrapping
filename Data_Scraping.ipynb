```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website to scrape
url = 'https://sugc.edu.bd/pages/notices'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the relevant data (e.g., notice titles)
    notices = soup.find_all('h5', class_='mb-1')

    # Find the relevant dates
    date_elements = soup.find_all("p", class_="mb-1 fs--1 text-500")

    # Extract text from the elements
    notices_text = [notice.get_text() for notice in notices]
    dates_text = [date.get_text() for date in date_elements]

    # Create a DataFrame to store the data
    df = pd.DataFrame({
        'Date': dates_text,
        'Notice': notices_text
    })
    # Save the DataFrame to a CSV file
    df.to_csv('notices.csv', index=False)
    print(df)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the website to scrape
base_url = 'https://sugc.edu.bd/pages/notices?page='

# Initialize lists to store the scraped data
all_notices = []
all_dates = []

# Loop through the first 50 pages
for page in range(1, 51):
    # Construct the URL for each page
    url = f"{base_url}{page}"

    # Send a GET request to the website
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find the relevant data (e.g., notice titles)
        notices = soup.find_all('h5', class_='mb-1')
        date_elements = soup.find_all("p", class_="mb-1 fs--1 text-500")

        # Extract text from the elements
        notices_text = [notice.get_text() for notice in notices]
        dates_text = [date.get_text() for date in date_elements]

        # Append the data to the lists
        all_notices.extend(notices_text)
        all_dates.extend(dates_text)
    else:
        print(f"Failed to retrieve data from page {page}: {response.status_code}")

# Create a DataFrame to store the data
df_college = pd.DataFrame({
    'Date': all_dates,
    'Notice': all_notices
})

# Save the DataFrame to a CSV file
df_college.to_csv('notices_sugc.csv', index=False)
# print(df)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the mobile catalog on Daraz Bangladesh
url = 'https://www.daraz.com.bd/catalog/?spm=a2a0e.tm80335411.search.d_go&q=mobile'

# Sending a request to the webpage
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)

# Parse the webpage content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all divs with the class "aBrP0" (where the price might be located)
divs = soup.find_all("div", class_="aBrP0")

# Loop through each div and extract the price
for div in divs:
    # Find the span tag with class "ooOxS" inside each div
    span_tag = div.find("span", class_="ooOxS")
    if span_tag:
        price = span_tag.get_text().strip()  # Use .get_text() to extract the text content
        print(price)
```
